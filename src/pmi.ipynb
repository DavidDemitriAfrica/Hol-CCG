```python
def is_number(s):
    try:
        float(s.replace(',', '').replace(' ', ''))
    except ValueError:
        return False
    else:
        return True
```

```python
corpus = []
N = 0
for i in range(0,24):
    i = str(i).zfill(2)
    with open(f'../../CCGbank/ccgbank_1_1/data/RAW/CCGbank.{i}.raw') as f:
        data = f.readlines()    
    for sentence in data:
        temp = []
        for word in sentence.split():
            if word not in [',','.']:
                if is_number(word):
                    word = '0'
                temp.append(word.lower())
        N += len(temp)
        corpus.append(' '.join(temp))
print(N)
```
